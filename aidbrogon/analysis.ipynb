## Analysis for 2020

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Load the data from Excel file
data = pd.read_excel('merge.xlsx')
```

```python
data.head()
```

```python
# Calculate overall turnover rate
data['Turnover Rate'] = (data['Leavers'] / data['Avg Headcount']) * 100
```

Top 10 Groups/Departments with Highest Turnover

```python
# Top 10 Departments with Highest Turnover
top_10_departments = data.head(10)
top_10_departments
```

Overall turnover rate of the organization

```python
(sum(data['Leavers']) / sum(data['Avg Headcount'])) * 100
```

Department-wise Turnover Comparison

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Group'], data['Turnover Rate'])
plt.xticks(rotation=90)
plt.xlabel('Department/Group')
plt.ylabel('Turnover Rate (%)')
plt.title('Department-wise Turnover Comparison')
plt.tight_layout()
plt.show()
```

Turnover and Average Headcount Visualization

```python
# Turnover and Average Headcount Visualization
plt.figure(figsize=(8, 5))
plt.scatter(data['Avg Headcount'], data['Turnover Rate'])
plt.xlabel('Average Headcount')
plt.ylabel('Turnover Rate (%)')
plt.title('Turnover vs. Average Headcount')
plt.tight_layout()
plt.show()
```

Turnover and Leavers Visualization

```python
# Turnover and Leavers Visualization
plt.figure(figsize=(8, 5))
plt.scatter(data['Leavers'], data['Turnover Rate'], color='red')
plt.xlabel('Leavers')
plt.ylabel('Turnover Rate (%)')
plt.title('Turnover vs. Leavers')
plt.tight_layout()
plt.show()
```

Correlation Analysis

```python
import seaborn as sns

correlation_matrix = data[['Leavers', 'Avg Headcount', 'Turnover Rate']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
